## 1. PDF to CSV

```python
def read_index_composition_from_pdf(base_path, file_name, export_path = True):
    #Open the PDF file in read-binary mode
    with open(base_path+file_name, 'rb') as f:
        pdf_reader = PyPDF2.PdfReader(f)
        slice_size = 5
        stocks = []
        for page_index in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_index]
            if page_index == 0:
                start_from = 7
            else:
                start_from = 6

            # Get the text of the page
            page_text = page.extract_text()
            # Split the text into lines
            lines = page_text.split('\n')[start_from:]

            for i in range(len(lines)):
                current_lines = lines[i*slice_size:(i+1)*slice_size]
                if len(current_lines) == 0: break
                stocks.append({'seq':current_lines[0],
                                        'ticker':current_lines[1].replace(' BS Equity','.SA'),
                                        'ponderacao':current_lines[2],
                                        'acoes':current_lines[3],
                                        'preco':current_lines[4]})
    stocks = pd.DataFrame(stocks)

    if export_path:
        stocks.to_csv(base_path+'/'+file_name.replace('.pdf','.csv'), index=False)

    return stocks
```

```python


# Transforming all pdf index to csv
base_path = '../data/ibov_comp_10_anos/'

for filename in os.listdir(folder_path):
    # check if the filename is a file (not a directory)
    if os.path.isfile(os.path.join(folder_path, filename)):
        read_index_composition_from_pdf(base_path, filename, True)
```

## 2. Downlaoding from yahoo finance the data - one download per index file

```python
# Transforming all pdf index to csv
base_path = '../data/ibov_comp_10_anos/'
export_prices_path = '../data/ibov_comp_10_anos/prices_v2/'

quarter_ibovespa = []
for filename in os.listdir(base_path):
    # check if the filename is a file (not a directory)
    if os.path.isfile(os.path.join(base_path, filename)):
        if '.csv' in filename:
            quarter_ibovespa.append(filename)
            
month = {'jan':'01', 'abr':'04','jul':'07', 'out':'10'}
backwards_historical, forward_historical = -365, 93
for index_file_name in quarter_ibovespa:
    _date = index_file_name[4:8]+'-'+ month[index_file_name[:3]]+'-01'
    backwards_date = dt.datetime.strptime(_date,'%Y-%m-%d') + dt.timedelta(days=backwards_historical)
    forward_date = dt.datetime.strptime(_date,'%Y-%m-%d') + dt.timedelta(days=forward_historical)
    tickers = list(pd.read_csv(base_path+index_file_name)['ticker'].values) + ['^BVSP']
    # downloading prices
    df_prices = data_utils.download_data(backwards_date, forward_date, tickers)
    # export
    df_prices.to_csv(export_prices_path+'prices_v2_'+index_file_name , index=False)
```
